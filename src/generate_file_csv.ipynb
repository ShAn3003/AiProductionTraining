```python
import os
import pandas as pd

def remove_dir_suffix(original_list):
    filtered_list = [element for element in original_list if not element.endswith("dir")]
    return filtered_list

def get_specify_suffix(original_list,str):
    filtered_list = [element for element in original_list if element.endswith(str)]
    return filtered_list

data_dir = '../Data'
file_dir_list = os.listdir(data_dir)
# print(file_dir_list)

target_list = []
label_list = []
for patient in file_dir_list:
    patient_root = os.path.join(data_dir,patient)
    # print(patient_root)
    study_lists = os.listdir(patient_root)
    study_lists = remove_dir_suffix(study_lists)
    # print(study_lists)
    for study in study_lists:
        study_root = os.path.join(patient_root,study)
        # print(study_root)
        series_lists = remove_dir_suffix(os.listdir(study_root))
        for series in series_lists:
            series_root = os.path.join(study_root,series)
            # print(series_root)
            target_lable_list = remove_dir_suffix(os.listdir(series_root))
            # print(target_lable_list)
            dcmi_lists = get_specify_suffix(target_lable_list,"dcm")
            xml_lists = get_specify_suffix(target_lable_list,"xml")
            # print(dcmi_lists)
            # print(xml_lists)
            for dcmi in dcmi_lists:
                dcmi_root = os.path.join(series_root,dcmi)
                index_d,_ = os.path.splitext(dcmi)
                index_d = int(index_d)
                xml_root = "-1"
                for xml in xml_lists:
                    index_x,_ = os.path.splitext(xml)
                    index_x = int(index_x)
                    if index_x ==index_d :
                        xml_root = os.path.join(series_root,xml)
                        break
                target_list.append(dcmi_root)
                label_list.append(xml_root)
                # break
            # break
        # break
    # break
original_df = pd.DataFrame({"target":target_list,"label":label_list})
original_df.to_csv("../datafile.csv",index=False)
```
